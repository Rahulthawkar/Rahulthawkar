# Exploratory Data Analysis on Car_Dekho Dataset

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### Load the Dataset

```python
Car_dekho= pd.read_csv('cardekho.csv')
Car_dekho
```

```python
Car_dekho.info()
```

```python
Car_dekho1=pd.DataFrame({'Unique value':Car_dekho.nunique(),
                'Missing value':Car_dekho.isnull().sum(),
                'Data Type': Car_dekho.dtypes})
Car_dekho1
```

```python
Car_dekho.columns
```

```python
Car_dekho['mileage(km/ltr/kg)'].isnull().sum()
```

```python
Car_dekho['mileage(km/ltr/kg)'].fillna(Car_dekho['mileage(km/ltr/kg)'].mean(), inplace=True)
```

```python
Car_dekho['engine'].fillna(Car_dekho['engine'].mean(), inplace=True)
```

```python
Car_dekho['max_power'] = Car_dekho['max_power'].str.replace('bhp', '', regex=False).str.strip()
Car_dekho['max_power'] = pd.to_numeric(Car_dekho['max_power'], errors='coerce')
```

```python
Car_dekho['max_power'].fillna(Car_dekho['max_power'].mean(), inplace=True)
```

```python
Car_dekho['seats'].fillna(Car_dekho['seats'].mode()[0], inplace=True)
```
