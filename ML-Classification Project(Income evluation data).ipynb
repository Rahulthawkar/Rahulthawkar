## Removing Outliers

```python
cols = ['age', 'workclass','education','education_num','martial_status','occupation','race','capital_gain','capital_loss','hrs_per_week','native_country']

Q1 = df[cols].quantile(0.25)
Q3 = df[cols].quantile(0.75)

IQR = Q3 - Q1

L_B = Q1 - 1.5 * IQR
U_B = Q3 + 1.5 * IQR

# Identify outilers : any column outside bounds 
outliers = (df[cols] < L_B) | (df[cols] > U_B)

ot = ~outliers.any(axis=1)


# filter the DataFrame

income  = df[ot].reset_index(drop=True)
```

# Hyperparameter Tuning

```python
from xgboost import XGBClassifier
```

```python
clf = XGBClassifier()
```

```python
from sklearn import metrics
from sklearn import model_selection
from sklearn import preprocessing
```

###  Defining a set of values as a dictionary for hyperparameters 

```python
param_grid = {
               "n_estimators" :  [100,200,300,400],
                "max_depth" : [1,3,5,7],
                "reg_lambda" : [0.01,0.1,0.5]
                }
```

###    Declaring GridSearchCV model

```python
model = model_selection.GridSearchCV(
        estimator = clf,
        param_grid = param_grid,
        scoring = 'accuracy',
        verbose = 10,
        n_jobs = 1,
        cv = 5
        )
```

### fitting values to the GridSearchCV model

```python
model.fit(x,y)
```

### Printing the best poosible values to enhance accuracy

```python
print(model.best_params_)
print()
```

```python
print(model.best_estimator_)
```

### printing the best score

```python
print(model.best_score_)
```
